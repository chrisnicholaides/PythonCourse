## visualisation

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('data/loan_data.csv')
```

## matplotlib

```python
fig, ax = plt.subplots(figsize=(8,4))

plt.scatter(df['Debt'], df['Balance'])
plt.title('Debt v Balance')
plt.xlabel('Debt')
plt.ylabel('Balance')


##ax.scatter(df['Debt'], df['Balance'])
```

## seaborn

```python
sns.scatterplot(data=df, x='Debt', y='Balance')
```
