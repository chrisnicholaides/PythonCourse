```python
import pandas as pd
import numpy as np
df = pd.read_csv('data/renfe_trains.csv')
```

## data inspection

```python
df =  df[df['price'] != 'price']
```

```python
df.loc[:,'price'] = df['price'].astype(float)
```

## missing values

```python
df.info()
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df[df['fare'].isna()]
```

```python
df.dropna(subset= ['fare', 'price', 'vehicle_class'],
         how='all',
         inplace=True)
```

```python
df.isna().sum()
```

```python
df[['vehicle_class', 'fare', 'price']].groupby(['vehicle_class', 'fare']).mean()
```

```python
df.fillna({'price' : df['price'].mean()}, inplace=True)
```

```python
df.isna().sum()
```
