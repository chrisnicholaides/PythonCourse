```python
df = pd.read_csv('data/loan_data.csv')
df
```

```python
df.drop('ID', axis=1, inplace=True)
```

```python
df
```

```python
df.replace({'Long Term' : '12 Months',
           'Short Term': '6 Months'})
```

```python
df['Debt']
```

```python
def debtbin(debt):
    if debt > 1000:
        return 'High'
    else:
        return 'low'
```

```python
debtbin(94)
```

```python
df['Debt'].map(debtbin)
```
