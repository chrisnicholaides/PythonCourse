## aggregating dataframes

```python
mortgage[['Balance','Term']].groupby('Term').mean()
```

```python
mortgage[['Income', 'Default']].groupby('Default').mean()
```

```python
mortgage[['Debt','Income','Balance', 'Default', 'Term']].groupby(['Default','Term']).mean()
```

```python
mortgage['MeanTermIncome'] = mortgage.groupby('Term')['Income'].transform('mean').round(2)
mortgage
```
